# Filtren los Pokémon que son de tipo 'Fighting' y tienen un Total mayor a 400.

```python
fighting = df[((df["Type 1"] == "Fighting") | (df["Type 2"] == "Fighting")) & (df["Total"] > 400)]
print(fighting)
```

```python
df["HP"] = df["HP"].astype(int)
```

```python
df = df.drop(columns=["Generation"])
```
