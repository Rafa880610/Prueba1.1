```python
df.replace(["UNKNOWN", "ERROR"], None, inplace=True)
```

## Homogeniza los valores de la columna ITEM, para que solo queden: [‘Coffee
## ‘Cake’, ‘Cookie’, ‘Salad’, ‘Smoothie’, ‘Sandwich’, ‘Juice’, ‘Tea’, None]`

```python
valores_permitidos = ['Coffee', 'Cake', 'Cookie', 'Salad', 'Smoothie', 'Sandwich', 'Juice', 'Tea']
```

```python
df['ITEM'] = df['ITEM'].apply(lambda x: x if x in valores_permitidos else None)
```

```python
df['ITEM'].unique()
```

## Homogeniza los valores de la columna PAYMENT_METHOD , para que solo queden:
## ['Credit Card', 'Cash', 'Digital Wallet', None]

```python
valores_permitidos_pago = ['Credit Card', 'Cash', 'Digital Wallet']
```

```python
df['PAYMENT_METHOD'] = df['PAYMENT_METHOD'].apply(lambda x: x if x in valores_permitidos_pago else None)
```

```python
df['PAYMENT_METHOD'].unique()
```

## Homogeniza los valores de la columna LOCATION , para que solo queden:
## ['Takeaway', 'In-store',None]

```python
valores_permitidos_ubicacion = ['Takeaway', 'In-store']
```

```python
df['LOCATION'] = df['LOCATION'].apply(lambda x: x if x in valores_permitidos_ubicacion else None)
```

```python
df['LOCATION'].unique()
```

## Convierte columnas a tipos de datos apropiados, como int , float , o datetime ,
## tomando en cuenta el manejo de errores.

```python
df['TOTAL_SPENT'] = pd.to_numeric(df['TOTAL_SPENT'], errors='coerce')
```

```python
df['QUANTITY'] = pd.to_numeric(df['QUANTITY'], errors='coerce')
```

```python
df['TRANSACTION_DATE'] = pd.to_datetime(df['TRANSACTION_DATE'], errors='coerce')
```

```python
df.dtypes
```

## Calcula las ventas totales por año utilizando groupby .

```python
df['YEAR'] = df['TRANSACTION_DATE'].dt.year
```

```python
df['YEAR'].unique()
```

```python
ventas_totales = df.groupby('YEAR')['TOTAL_SPENT'].sum()
```

```python
ventas_totales
```

## ¿Cuáles son los meses con mayor venta?

```python
df['YEAR_MONTH'] = df['TRANSACTION_DATE'].dt.to_period('M')
```

```python
ventas_por_mes = df.groupby('YEAR_MONTH')['TOTAL_SPENT'].sum()
```

```python
ventas_por_mes = ventas_por_mes.sort_values(ascending=False)
```

```python
ventas_por_mes
```

## Identifica los productos más vendido

```python
productos_mas_vendidos = df.groupby('ITEM')['QUANTITY'].sum().sort_values(ascending=False)
```

```python
productos_mas_vendidos
```

## De los 3 productos más vendidos, ¿Cuál es el más vendido por mes ?

```python
df['MONTH'] = df['TRANSACTION_DATE'].dt.month
```

```python
top_3_productos = df.groupby('ITEM')['QUANTITY'].sum().nlargest(3).index
```

```python
df_top = df[df['ITEM'].isin(top_3_productos)]
```

```python
ventas_por_mes = df_top.groupby(['MONTH', 'ITEM'])['QUANTITY'].sum().reset_index()
```

```python
producto_mas_vendido_por_mes = ventas_por_mes.loc[ventas_por_mes.groupby('MONTH')['QUANTITY'].idxmax()]
```

```python
producto_mas_vendido_por_mes
```
