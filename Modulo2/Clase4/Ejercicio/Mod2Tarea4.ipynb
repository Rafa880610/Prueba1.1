# Transforma el titulo de las peliculas a mayusculas, reemplazando espacios con guiones bajos

```python
df.loc[df["type"] == "Movie", "title"] = df["title"].str.upper().str.replace(" ", "_")
```

# Crea una columna llamada duración

```python
df.loc[df["type"] == "Movie", "duración"] = df["duration"].apply(lambda x: "corta" if x < 120 else "larga")
```

# Calcula

```python
df["year_added"] = df["date_added"].dt.year
```

```python
df["main_genre"] = df["listed_in"].str.split(",").str[0]
```

```python
grouped = df.groupby(["year_added", "main_genre"]).agg(
    count_titles=("title", "count"),
    avg_duration=("duration", "mean")
).reset_index()
```

```python
grouped["percent_titles"] = grouped.groupby("year_added")["count_titles"].transform(lambda x: (x / x.sum()) * 100)
```

```python
grouped
```

# Ordenar

```python
grouped = grouped.sort_values(by=["year_added", "main_genre"], ascending=[False, False])
```
